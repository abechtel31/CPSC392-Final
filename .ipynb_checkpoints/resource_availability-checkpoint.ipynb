## Community Health Equity Lab: New York COVID19 Response Analysis

### QUESTION Does the strictness with which neighborhoods follow social distancing predict COVID cases?

```python
import geopandas as gpd
import folium
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split # simple TT split cv
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

pd.options.display.max_columns = 300
pd.set_option('display.max_rows', 500)
%matplotlib inline
```

## Data

```python
ny = pd.read_csv('output/ny_final.csv')
ny.head()
```

```python
ny_shp = gpd.read_file('data/spatial/bor_zip_codes.shp')
```

```python
ny = pd.merge(ny_shp, ny, how='inner', left_on='GEOID10', right_on='GEOID_ZIP')
ny.crs= {'init': 'epsg:4326'} #WGS84
```

### Visualizations

### Modeling

```python
predictors = ["med_income", "pcnt_pov", "minority-majority-50", "4_16_2020_positive"]

X_train, X_test, y_train, y_test = train_test_split(ny[predictors], ny["ICU_beds"], test_size=0.2)
```

```python
mod = sm.OLS(y_train,X_train)
fit = mod.fit()
```

```python
fit.summary()
```

```python
test_preds = fit.predict(X_test)
```

```python
print('testing r2 is:', r2_score(test_preds, y_test))
```
