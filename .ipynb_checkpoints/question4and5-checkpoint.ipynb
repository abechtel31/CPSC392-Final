```python
#standardize variables
zScore = StandardScaler() #standardize variables makes them easier with the math
zScore.fit(df[features])
df[features] = zScore.transform(df[features])
```

```python
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer(n_factors = 4, rotation=None)
fa.fit(df[features])

loadings = fa.loadings_

# Check Eigenvalues
ev, v =  fa.get_eigenvalues()

#create dataframe with relations between factors and variables
meaning = pd.DataFrame(loadings, columns = factors, index=features)
meaning
```

```python
# get the column name of max values in every row
maxValueIndexObj = meaning.idxmax(axis=1)
 
print("Max values of row are at following columns :")
print(maxValueIndexObj)
```

```python
xvals = range(1,df[features].shape[1]+1)
```

```python
# Create scree plot using matplotlib
plt.scatter(xvals,ev)
plt.plot(xvals,ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

```python
variance = fa.get_factor_variance()

vDF = pd.DataFrame(variance, columns = ['factor1', 'factor2','factor3','factor4'], index= ["SS Loadings", "Proportion Var", "Cumulative Var"])
vDF
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[features], df["per_infected"], test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

train_pred = model.predict(X_train)
test_pred = model.predict(X_test)


print('training r2 is:', model.score(X_train, y_train)) #training R2
print('testing r2 is:', model.score(X_test, y_test)) #testing R2

print('\ntrain mse is: ', mean_squared_error(y_train,train_pred))
print('test mse is: ', mean_squared_error(y_test,test_pred))
```

```python
true_v_pred = pd.DataFrame({"predict": test_pred, "trueV": y_test})
ggplot(true_v_pred, aes(x = "trueV", y = "predict")) +geom_point() +theme_minimal()
```

```python
coefficients = pd.DataFrame({"Coef":model.coef_,
              "Name": features})

coefficients = coefficients.append({"Coef": model.intercept_,
               "Name": "intercept"}, ignore_index = True)

print(coefficients)
```

```python
X = ny[["tot_pop", "male", "female", "pop_density", "per_minority"]]
y = ny["per_infected"]

zScore = StandardScaler() #standardize variables makes them easier with the math
zScore.fit(X)
X = zScore.transform(X)

model = kb.Sequential()
model.add(kb.layers.Dense(10, input_shape=[5], activation='relu'))
model.add(kb.layers.Dense(5, activation='relu'))
model.add(kb.layers.Dense(1, activation='sigmoid'))

# compile the keras model
model.compile(loss="mean_squared_error", optimizer='adam', metrics=['accuracy'])

# fit the keras model on the dataset
model.fit(X, y, epochs=150, batch_size=10)

# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
